```python
import numpy as np
import pandas as pd

df_dataset = pd.read_csv('student-mat.csv', sep=',', index_col=None)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.utils import shuffle

df_dataset = shuffle(df_dataset, random_state=42)

target = 'sex'
X_columns = df_dataset.columns.tolist()
X_columns.remove(target)
X = df_dataset[X_columns]
y = df_dataset[target]

y_binary = pd.get_dummies(y, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=200)

categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()

X_train = pd.get_dummies(X_train, columns=categorical_cols)
X_test = pd.get_dummies(X_test, columns=categorical_cols)

X_train_array = X_train.values

normalizer = preprocessing.Normalization()
normalizer.adapt(X_train_array)

model = keras.Sequential([
    normalizer,
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train_array, y_train, epochs=40, validation_data=(X_test, y_test))

def plot_loss(history):
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.xlabel('Epochs')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(True)

plot_loss(history)

y_pred_proba = model.predict(X_test)
y_pred = (y_pred_proba > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_report_str = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(classification_report_str)

plt.scatter(y_test, y_pred_proba)
plt.xlabel('Rótulos Reais')
plt.ylabel('Probabilidades Previstas')
plt.title('Rótulos Reais vs. Probabilidades Previstas')
plt.show()
```
